<h1>PERSONALITY PREDICTION THROUGH CV ANALYSIS</h1>

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('training_dataset.csv')
```

```python
df.head()
```

<H1>PRE PROCESSING</H1>

```python
df.describe()
```

```python
df.columns
```

```python
df=df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['Gender']=le.fit_transform(df['Gender'])
df['Personality (Class label)']=le.fit_transform(df['Personality (Class label)'])
```

```python
df
```

<H1>MODEL FITTING</H1>

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
pred=lr.predict(xtest)
```

```python
acc=accuracy_score(pred,ytest)
print(acc)
```

```python
import tkinter as tk
from tkinter import ttk, filedialog

class PersonalityForm(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Personality Form")
        self.configure(bg='#006400')  
        self.create_widgets()

    def predict(self):
        # Get user input from the GUI
        applicant_name = self.name_entry.get()
        age = int(self.age_entry.get())
        gender = self.gender_var.get()
        openness = self.sliders[0].get()
        neuroticism = self.sliders[1].get()
        conscientiousness = self.sliders[2].get()
        agreeableness = self.sliders[3].get()
        extraversion = self.sliders[4].get()

        # Example: Create a feature vector based on user input
        features = [age, openness, neuroticism, conscientiousness, agreeableness, extraversion]

        # Add gender as a numeric feature (assuming Male=0, Female=1)
        gender_numeric = 0 if gender == "Male" else 1
        features.insert(0, gender_numeric)

        # Example: Make a prediction using your machine learning model
        prediction = lr.predict([features])

        personality_traits = self.convert_to_personality_traits(prediction[0])

        # Display the prediction on the GUI
        ttk.Label(self, text=f"Prediction for {applicant_name}: {personality_traits}", background="#006400", font=(10), foreground="#fff").grid(row=9, column=0, padx=5, pady=5, sticky="w")

    
    def convert_to_personality_traits(self, prediction):
        # Convert numeric prediction to specific personality traits
        traits_dict = {0: "Openess", 1: "neuroticism", 2: "conscientiuusness", 3: "agreeableness", 4: "Extraversion"}
        return traits_dict.get(prediction, "Unknown Trait")


        # Example: Display the prediction (you may want to update this based on your application)
        print(f"Prediction for {applicant_name}: {prediction}")

    def create_widgets(self):
        # Create a frame to hold the form widgets
        form_frame = ttk.Frame(self, padding=(20, 10), style="My.TFrame")
        form_frame.grid(row=0, column=0, sticky="nsew")

        # Apply a custom style to the frame
        self.style = ttk.Style(self)
        self.style.configure("My.TFrame", background='#006400')  

        # Applicant Name
        ttk.Label(form_frame, text="Applicant Name:", background="#006400", font=(10), foreground="#fff", style="Bold.TLabel").grid(row=0, column=0, padx=5, pady=5, sticky="w")
        self.name_entry = ttk.Entry(form_frame)
        self.name_entry.grid(row=0, column=1, padx=5, pady=5, sticky="w")

        # Age
        ttk.Label(form_frame, text="Age:", background="#006400", font=(10), foreground="#fff", style="Bold.TLabel").grid(row=1, column=0, padx=5, pady=5, sticky="w")
        self.age_entry = ttk.Entry(form_frame)
        self.age_entry.grid(row=1, column=1, padx=5, pady=5, sticky="w")
        
        # Gender
        ttk.Label(form_frame, text="Gender:", background="#006400", font=(10), foreground="#fff", style="Bold.TLabel").grid(row=2, column=0, padx=5, pady=5, sticky="w")
        self.gender_var = tk.StringVar()
        self.gender_var.set("Male")
        ttk.Radiobutton(form_frame, text="Male", variable=self.gender_var, value="Male", style="My.TRadiobutton").grid(row=2, column=1, padx=5, pady=5, sticky="w")
        ttk.Radiobutton(form_frame, text="Female", variable=self.gender_var, value="Female", style="My.TRadiobutton").grid(row=2, column=2, padx=5, pady=5, sticky="w")

        # Resume Upload
        ttk.Label(form_frame, text="Upload Resume:", background="#006400", font=(10), foreground="#fff", style="Bold.TLabel").grid(row=3, column=0, padx=5, pady=5, sticky="w")
        self.upload_button = ttk.Button(form_frame, text="Browse",style="My.TButton", command=self.browse_resume)
        self.upload_button.grid(row=3, column=1, padx=5, pady=5, sticky="w")

        # Personality Traits
        traits_frame = ttk.LabelFrame(form_frame, text="Personality Traits", padding=(10, 5), style="My.TLabelframe")

        # Personality Traits traits_frame = ttk.LabelFrame(form_frame, text="Personality Traits", padding=(10, 5), style="My.TLabelframe")
        traits_frame.grid(row=4, column=0, columnspan=3, pady=10, sticky="nsew")

        # Apply a custom style to the labeled frame
        self.style.configure("My.TLabelframe", background='#008000', forground="#fff")  

        labels = ["Openness", "Neuroticism", "Conscientiousness", "Agreeableness", "Extraversion"]
        self.sliders = [ttk.Scale(traits_frame, from_=0, to=10, orient="horizontal") for _ in range(5)]

        for i, label in enumerate(labels):
            ttk.Label(traits_frame, text=f"{label}:", background="#008000", font=(10), foreground="#fff", style="Bold.TLabel").grid(row=i, column=0, padx=5, pady=5, sticky="w")
            self.sliders[i].grid(row=i, column=1, padx=5, pady=5, sticky="w")
            
        self.submit_button = ttk.Button(form_frame, text="Submit", style="My.TButton", command=self.predict)
        self.submit_button.grid(row=8, column=0, columnspan=3, padx=5, pady=10, sticky="w")
        
        # Apply a custom style to the button
        self.style.configure("My.TButton", background='#00008B', foreground='#fff', padding=(10, 5))
        
        # Apply a custom style to the radio button
        self.style.configure("My.TRadiobutton", background='#006400',foreground="#fff",font=(8))

    def browse_resume(self):
        file_path = filedialog.askopenfilename(title="Select Resume", filetypes=[("PDF Files", "*.pdf"), ("Text Files", "*.txt"), ("All Files", "*.*")])
        # You can use the 'file_path' as needed, for example, display the file name or store it for submission.

if __name__ == "__main__":
    form = PersonalityForm()
    form.mainloop()
```
